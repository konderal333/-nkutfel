```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# I didn't want to use a 'for' cycle, because 5 pictures isn't a lot

# 1st image

im1 = Image.open(r"images/image1.png") 

print("First image:")
print("")

display(im1)                                     # displaying the image

print("")
print("Histograms of the colour channels:")

im1_array = np.asarray(im1)                      # getting the image as a numpy array

red1 = im1_array[:,:,0].flatten()
green1 = im1_array[:,:,1].flatten()               #obtaining and flattening the arrays of the colour channels
blue1 = im1_array[:,:,2].flatten()

fig1, axes1 = plt.subplots(1, 3,figsize=(15,6), sharey=True)
axes1[0].hist(red1, bins=256, color='r');                # plotting the histograms
axes1[1].hist(green1, bins=256, color='g');
axes1[2].hist(blue1, bins=256);
plt.show()

print("----------------------------------------------------------------------------------------------------------")

# 2nd image

im2 = Image.open(r"images/image2.png") 

print("Second image:")
print("")

display(im2)                                     # displaying the image

print("")
print("Histograms of the colour channels:")

im2_array = np.asarray(im2)                      # getting the image as a numpy array

red2 = im2_array[:,:,0].flatten()
green2 = im2_array[:,:,1].flatten()               #obtaining and flattening the arrays of the colour channels
blue2 = im2_array[:,:,2].flatten()

fig2, axes2 = plt.subplots(1, 3,figsize=(15,6), sharey=True)
axes2[0].hist(red2, bins=256, color='r');                # plotting the histograms
axes2[1].hist(green2, bins=256, color='g');
axes2[2].hist(blue2, bins=256);
plt.show()

print("----------------------------------------------------------------------------------------------------------")

# 3rd image

im3 = Image.open(r"images/image3.jpg") 

print("Third image:")
print("")

display(im3)                                     # displaying the image

print("")
print("Histograms of the colour channels:")

im3_array = np.asarray(im3)                      # getting the image as a numpy array

red3 = im3_array[:,:,0].flatten()
green3 = im3_array[:,:,1].flatten()               #obtaining and flattening the arrays of the colour channels
blue3 = im3_array[:,:,2].flatten()

fig3, axes3 = plt.subplots(1, 3,figsize=(15,6), sharey=True)
axes3[0].hist(red3, bins=256, color='r');                # plotting the histograms
axes3[1].hist(green3, bins=256, color='g');
axes3[2].hist(blue3, bins=256);
plt.show()

print("----------------------------------------------------------------------------------------------------------")

# 4th image

im4 = Image.open(r"images/image4.jpeg") 

print("Fourth image:")
print("")

display(im4)                                     # displaying the image

print("")
print("Histograms of the colour channels:")

im4_array = np.asarray(im4)                      # getting the image as a numpy array

red4 = im4_array[:,:,0].flatten()
green4 = im4_array[:,:,1].flatten()               #obtaining and flattening the arrays of the colour channels
blue4 = im4_array[:,:,2].flatten()

fig4, axes4 = plt.subplots(1, 3,figsize=(15,6), sharey=True)
axes4[0].hist(red4, bins=256, color='r');                # plotting the histograms
axes4[1].hist(green4, bins=256, color='g');
axes4[2].hist(blue4, bins=256);
plt.show()

print("----------------------------------------------------------------------------------------------------------")

# 5th image

im5 = Image.open(r"images/image5.jpg") 

print("Fifth image:")
print("")

display(im5)                                     # displaying the image

print("")
print("Histograms of the colour channels:")

im5_array = np.asarray(im5)                      # getting the image as a numpy array

red5 = im5_array[:,:,0].flatten()
green5 = im5_array[:,:,1].flatten()               #obtaining and flattening the arrays of the colour channels
blue5 = im1_array[:,:,2].flatten()

fig5, axes5 = plt.subplots(1, 3,figsize=(15,6), sharey=True)
axes5[0].hist(red5, bins=256, color='r');                # plotting the histograms
axes5[1].hist(green5, bins=256, color='g');
axes5[2].hist(blue5, bins=256);
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# the scaler to standardize
scaler = StandardScaler()

# 1st image

print("For the first image:")
print("")
print("Before scaling:")
print("- Red channel's mean:".ljust(27), np.mean(red1), ",\t variance:", np.std(red1))
print("- Green channel's mean:".ljust(27), np.mean(green1),",\t variance:", np.std(green1))
print("- Blue channel's mean:".ljust(27), np.mean(blue1), ",\t variance:", np.std(blue1))
print("")


red1_scaled = scaler.fit_transform(red1.reshape(-1, 1))
green1_scaled = scaler.fit_transform(green1.reshape(-1, 1))        # we have to reshape the data in order to scale it 
blue1_scaled = scaler.fit_transform(blue1.reshape(-1, 1))

print("After scaling:")
print("- Red channel's mean:".ljust(27), np.mean(red1_scaled), ",\t variance:", np.std(red1_scaled))
print("- Green channel's mean:".ljust(27), np.mean(green1_scaled),", \t variance:", np.std(green1_scaled))
print("- Blue channel's mean:".ljust(27), np.mean(blue1_scaled), ", \t variance:", np.std(blue1_scaled))
print("--------------------------------------------------------------------------------------------------")

# TODO: maradk
```
